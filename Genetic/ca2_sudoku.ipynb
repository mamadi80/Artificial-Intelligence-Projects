In this assigmnet , we will find the answer for a 9 by 9 sudoku using genetic algorithm .<br>
We have an input sudoko which every member of population must be same with it in non zero cells <br>
, so we will reset this cells in every new child too .<br>
In terms of population I chose 1000 members and I got to this number by trial and error ,<br>
cause if the population gets really small , it would be harder to make random new better sudokos ,<br>
and if our population is too big we will need more time to check all members of this big population<br>
but there is more chance to make new better childs,<br>
so we will lose time and that's a trade-off .<br>
<br>
Definition of some parameters in this assignment :<br>
<br>
Gene : We choose each cell of sudoku table as a gene here , so 81 of them together make a sudoku table.<br>
<br>
chromosome : Each sudoku in our list of sudokos is a chromosome , and we are looking for one with<br>
the traits we want .<br>
And here we want to find a sudoku which has no repetitious number in any row, column and 3 by 3 block  <br>
which also has the same cells with input sudoku.<br>
<br>
Population : It has been described before , but to mention it's first made by random sudoko tables<br>
in amount of 1000 (total) , and it changes every generation by choosing all the childs made by parents <br>
which will make 500 (total/2) new members , and the other half will be choose randomly of the old population we have .<br>
<br>
Also for choosing this half and the parents to make childs (selection) , <br>
I use numpy random choice with chances given based on each sudoku fitness ,<br>
so the members with more fitness will have more chance to get selected for making childs and going to next generation .<br>
<br>
Fitness : I will give each sudoku a number of fitness which shows better fitness for higher numbers <br>
          and is equal to :<br>
          216 - sum ( (repeats of each number in a row - 1) + (repeats of each number in a column - 1) + (repeats of each number in a 3 by 3 block - 1) )<br>
          This number can be between 0  (if all cells are the same number) and 216 (if we solve the problem) <br>
          and generally is decided by amount of wrong number placements in sudoku .<br>
          <br>
chance of being chosen : It is used in our numpy random choice and for each sudoku is equal to<br>
                         fitness_of_this_sudoku / sum_of_all_fitnesses . <br>
                         So members with higher fitness are more likely to be chosen . (roulette wheel selection) <br>
                         <br>
Crossover : The way we make a new child from two selected parents . In this assigment I tried the same way <br>
            as mentioned in class , and I will make first row of child from first row of parent1 ,<br>
            second row from second row of parent2 , third row from third row of parent1 and ... till all 9 rows are made.<br>
            <br>
Mutation (and percentage) : Each child that is created , has a 20% chance of mutation , and if it happens<br>
           one random row of this child will be replaced by a random new list .<br>
           It will help us not being stuck in local minima and have a better chance of getting close to right answer.<br>
           This percentage is choosen after trial and error .<br>
Crossover percentage : 50% (half of our new population is made of all new childs , and half is choosen of the generation before them).<br>
                       It is a good number cause we wouln't lose much of our old population which might have some fit members ,<br>
                       but we also don't get stuck and choose from new generation which can mutate and help us from being<br>
                       stuck too.<br>
As we can see , our improvement gets slower the closer we get to right answer , because our generation gets alike more which would<br>
lead to alike children too , and also we have less right places of numbers to find , so our general chance of improvement changes too .<br>
Mutation and ways like lowering the chance of selecting old generation when we are getting stopped , can help us out of this situation .<br>
