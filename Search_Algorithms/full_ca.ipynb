My state model include : map , valued by color values of our map (that shows blocked cells , potion cells and double cells)
                        and doctors positions , that we decide by this two the uniqueness of our state 
                        for IDS , we have also the length of our path cause we can get to a state with diffrent number of moves for diffrent lengths
                        
Goal state : We finish our algorithm when all doctors are in goal state and all potions are collected (collected potions number = c)

Initial state : We start our algorithm by a doctor in cell[0][0] , number of potions = 0 and zero moves , with the table made by colors of input

Action : For every doctor in any cell , we can have at most 4 new cells to check , which some might be considered repetitious , and some might be not possible to move into
         so we add the cells that have our conditions to end of our queue
         
Queue : We add every possible move to our queue and do the first move (queue[0]) , then search for new possible moves . Until we see all possible states to get to our goal

BFS : Breadth first search . we add every cell 1 move away from our current cell . It needs less amount of time than IDS and DFS  , But more amount of memory to save our states .

IDS : Iterative deepening search , We start from our first state and max length of 0 , and do DFS till we reach our max lenghth . after we are done with our dfs , we add 
      1 unit to our max length and do DFS again like before . We keep adding max length till we find our path to goal . It needs more time than two others , But less memory than 

Notice that all of our algorithms expect weighted A* find the optimal path .

In terms of memory : A* better than IDS better than BFS .
In terms of time complexity : Weighted A* better than A* better than BFS better than IDS .
